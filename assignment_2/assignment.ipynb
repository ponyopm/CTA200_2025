```python
# Part 1:

def f(x):
    """
    A polynomial x^3 - x^2 -1.
    
    """
    return x**3 - x**2 - 1

def df(x):
    """
    The derivative of the function f with respect to x.
    
    """
    return 3*x**2 -2*x
```

```python
# Part 2:

def newton(f, df, x0, epsilon=1e-6, max_iter=30):
    iter = 0
    x = x0
    while iter < max_iter:
        if abs(f(x)) < epsilon:
            print(f"Found root in {iter} iterations.")
            return x
        else:
            x = x - (f(x)/df(x))
            iter +=1
    print(f"Iteration failed")
    return None
```

```python
# Part 3:

newton(f, df, 10, epsilon=1e-6, max_iter=30)
```
